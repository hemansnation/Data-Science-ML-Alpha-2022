# Group By

```python
import pandas as pd
```

```python
df = pd.read_csv('weather_by_cities.csv')
df
```

```python
gb = df.groupby('city')
gb
```

```python
for city, data in gb:
    print("City :",city)
    print()
    print(data)
```

```python
gb.get_group('mumbai')
```

```python
gb.max()
```

```python
gb.mean()
```

```python
gb.describe()
```

```python
gb.size()
```

```python
gb.count()
```

# Concatenation

```python
indian_weather = pd.DataFrame({
    "city": ['mumbai', 'delhi', 'banglore'],
    "temperature": [32,45,30],
    "humidity": [80,60,78]
})
indian_weather
```

```python
us_weather = pd.DataFrame({
    "city": ['new york', 'chicago', 'orlando'],
    "temperature": [21,14,25],
    "humidity": [88,25,75]
})
us_weather
```

```python
df = pd.concat([indian_weather, us_weather])
df
```

```python
df = pd.concat([indian_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([indian_weather, us_weather], keys= ["india", "us"])
df
```

```python
df.loc["us"]
```

```python
type(df.loc["us"])
```

```python
df = pd.concat([indian_weather, us_weather], axis=1)
df
```

```python
s = pd.Series(["Humid", "Dry", "Rain"], name="Event")
s
```

```python
df = pd.concat([indian_weather, s], axis=1)
df
```
