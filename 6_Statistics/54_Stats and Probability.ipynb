```python
import random

import numpy as np
import pandas as pd

import scipy.stats

import seaborn as sns
import matplotlib.pyplot as plt
```

# Coin

```python
def flip(n):
    
    result = []
    
    for i in range(1, n + 1):
        result.append(random.choice(['Head', 'Tails']))
    
    return result
```

```python
f = flip(10)
f
```

```python
plt.figure(figsize=(12,8))

sns.countplot(x = f)

plt.xlabel('Coin Flips')

plt.title('Counts')

plt.show()
```

```python
f = flip(100)

plt.figure(figsize=(12,8))

sns.countplot(x = f)

plt.xlabel('Coin Flips')

plt.title('Counts')

plt.show()
```

```python
f = flip(100000)

plt.figure(figsize=(12,8))

sns.countplot(x = f)

plt.xlabel('Coin Flips')

plt.title('Counts')

plt.show()
```

# Dice

```python
def roll(n):
    
    result = []
    
    for i in range(1, n + 1):
        result.append(random.choice([1,2,3,4,5,6]))
    
    return result
```

```python
r = roll(10)
r
```

```python
f = roll(10)

plt.figure(figsize=(12,8))

sns.countplot(x = f)

plt.xlabel('Coin Flips')

plt.title('Counts')

plt.show()
```

```python
f = roll(100)

plt.figure(figsize=(12,8))

sns.countplot(x = f)

plt.xlabel('Coin Flips')

plt.title('Counts')

plt.show()
```

```python
f = roll(100000)

plt.figure(figsize=(12,8))

sns.countplot(x = f)

plt.xlabel('Coin Flips')

plt.title('Counts')

plt.show()
```
