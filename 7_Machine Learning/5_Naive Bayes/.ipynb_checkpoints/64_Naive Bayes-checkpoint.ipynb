# Naive Bayes

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# gaussian naive bayes - when your data follows normal distribution/gaussian
                            # distribution
```

```python
from sklearn.naive_bayes import GaussianNB # classification

gb = GaussianNB()

gb.fit(df, iris.target)
```

```python
gb.score(df,iris.target)
```

```python
gb.predict([[5.1,3.5,1.4,0.2]])
```

# Datasets

```python
docs = pd.read_table('SMSSpamCollection+(1)', 
                     header=None, 
                     names=['Class', 'SMS'])

docs.head()
```

```python
ham_spam = docs.Class.value_counts()
ham_spam
```

```python
docs['label'] = docs.Class.map({ 'ham':0, 'spam':1 })

docs.head()
```

```python
x = docs.SMS
y = docs.label

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=101)
```

```python
xtrain.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(stop_words='english')
```

```python
vect.fit(xtrain)
```

```python
vect.vocabulary_
```

```python
xtrain_transformed = vect.transform(xtrain)
xtest_transformed = vect.transform(xtest)
```

```python
print(type(xtrain_transformed))
print(xtrain_transformed)
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()

mnb.fit(xtrain_transformed, ytrain)

ypred_class = mnb.predict(xtest_transformed)

ypred_proba = mnb.predict_proba(xtest_transformed)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest,ypred_class)
```

```python
ypred_class
```

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(ytest,ypred_class)
```
