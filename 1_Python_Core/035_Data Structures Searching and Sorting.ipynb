## Linear Search

```python
n = [11,23,45,33,77,54,99,66]
#     0  1  2  3  4  5  6  7
#    k

k = int(input("Enter a number :"))

for i in range(len(n)):
    if n[i] == k:
        print("Search Success")
        break
```

```python
n = [11,23,45,33,77,54,99,66]
#     0  1  2  3  4  5  6  7
#    k

k = int(input("Enter a number :"))

#flag bit variable
f = 0

for i in range(len(n)):
    if n[i] == k:
        f = 1
        break

if f == 1:
    print("Search Success")
else:
    print("Search Fail")
```

# Swapping

```python
a = 10  # a = 10, 20
b = 20  # b = 20, 10
        # c = 10, 
c = a
a = b
b = c

print(a)
print(b)
```

#### swapping without using 3rd variable

```python
a = 10  # a = 10, 30, 20
b = 20  # b = 20, 10

a = a + b
b = a - b
a = a - b

print(a)
print(b)
```

### swapping with tuple unpacking

```python
a = 10
b = 20

a,b = b,a
#a,b = (20,10)

print(a)
print(b)
```

# Sorting

```python
  [40,50,10,20,30]
#        i  j
#
```

```python
n = [10,50,40,20,30]
#    0  1  2  3  4
#    i        j
i = 0

#     0       4
while i < len(n) - 1:
    j = i + 1
    while j < len(n):
        if n[i] > n[j]:
            temp = n[i]
            n[i] = n[j]
            n[j] = temp
        j += 1
    i += 1

print("Sorted List: ", n)
```
