# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [50, 51,52,48,47,49,46]
```

```python
plt.xlabel("Day")
```

```python
plt.ylabel('Temperature')
```

```python
plt.title("Weather")
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y)
```

```python
plt.plot(x,y, color='red', linewidth = 3)
```

```python
plt.plot(x,y, color='red', linewidth = 3, linestyle='dotted')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color='red', linewidth = 3, linestyle='dashed')
```

# Format strings

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'r--')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color='blue', linewidth = 2, marker='*' , linestyle='dotted')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'r+--')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'g^')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'g<')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'g<', alpha=0.6)   # alpha = opacity - rgba 
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color='red', marker='+', linestyle='')
```

# Label parameter

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend(loc='best')
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend(loc='upper right', shadow=True, fontsize='large')
```
